# **Outlier Detection**

```python
# Select the numerical columns
numerical_columns = ["Age", "Education", "Usage", "Fitness", "Income", "Miles"]

# Detect outliers using IQR method
outliers = {}
for col in numerical_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers[col] = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

# Display outliers
for col, outlier_data in outliers.items():
    if not outlier_data.empty:
        print(f"Outliers in {col}:\n{outlier_data}\n")
    else:
        print(f"No outliers in {col}.")
```

# **Conditional Probabilities**

What percent of customers have purchased KP281, KP481, or KP781?

```python
frequency = df['Product'].value_counts()
print(frequency)
percentage = (frequency / len(df)) * 100
percentage
```

Create frequency tables and calculate the percentage as follows

**o Product - Gender**

Percentage of a Male customer purchasing a treadmill

Percentage of a Female customer purchasing KP781 treadmill

Probability of a customer being a Female given that Product is
KP281

```python
# 1. Percentage of Male customers purchasing a treadmill
male_treadmill_percentage = (df[df['Gender'] == 'Male'].shape[0] / df.shape[0]) * 100

# 2. Percentage of Female customers purchasing KP781 treadmill
kp781_female_percentage = (df[(df['Product'] == 'KP781') & (df['Gender'] == 'Female')].shape[0] /df[df['Product'] == 'KP781'].shape[0]) * 100

# 3. Probability of a customer being Female given that the product is KP281
kp281_female_probability = (df[(df['Product'] == 'KP281') & (df['Gender'] == 'Female')].shape[0] /df[df['Product'] == 'KP281'].shape[0])

print(f"Percentage of Male customers purchasing a treadmill: {male_treadmill_percentage:.2f}%")
print(f"Percentage of Female customers purchasing KP781: {kp781_female_percentage:.2f}%")
print(f"Probability of a customer being Female given Product is KP281: {kp281_female_probability:.4f}")
```

**Product - Age**

Percentage of customers with Age between 20s and 30s among all
customers

```python
length_customers = len(df["Gender"])
age_20_30 = len(df[(df["Age"] >= 20) & (df["Age"] <= 30)])

perc_age_20_30 = ((age_20_30)/(length_customers)) * 100

print(f"Percentage of customers with Age between 20s and 30s among all customers: {perc_age_20_30:.2f}%")
```

**Product - Income**

Percentage of a low-income customer purchasing a treadmill

Percentage of a high-income customer purchasing KP781
treadmill

Percentage of customer with high-income salary buying treadmill
given that Product is KP781

```python
# Define thresholds for low and high income
low_income_threshold = 40000
high_income_threshold = 80000

# Filter low-income and high-income customers
low_income_customers = df[df['Income'] < low_income_threshold]
high_income_customers = df[df['Income'] > high_income_threshold]

# 1. Percentage of low-income customers purchasing a treadmill
low_income_treadmill_percentage = (low_income_customers.shape[0] / df.shape[0]) * 100

# 2. Percentage of high-income customers purchasing KP781
high_income_kp781_percentage = (high_income_customers[high_income_customers['Product'] == 'KP781'].shape[0] / high_income_customers.shape[0]) * 100

# 3. Percentage of high-income customers buying a treadmill given Product is KP781
kp781_customers = df[df['Product'] == 'KP781']
high_income_given_kp781_percentage = (kp781_customers[kp781_customers['Income'] > high_income_threshold].shape[0] / kp781_customers.shape[0]) * 100

print(f"Percentage of low-income customers purchasing a treadmill: {low_income_treadmill_percentage:.2f}%")
print(f"Percentage of high-income customers purchasing KP781: {high_income_kp781_percentage:.2f}%")
print(f"Percentage of high-income customers buying a treadmill given Product is KP781: {high_income_given_kp781_percentage:.2f}%")
```

Percentage of customer with fitness level 5 buying KP781

# **Actionable Insights & Recommendations**

### **Product Analysis**
1. **KP281** is favored by younger users with the lowest average age among the products.  
   - **Recommendation**: Market KP281 as a trendy and affordable option for younger customers through youth-centric advertising channels (e.g., social media platforms).  

2. **KP481** appeals to a middle-aged demographic.  
   - **Recommendation**: Emphasize its balanced features in promotional material, targeting professionals and families in the 30–40 age range.  

3. **KP781** caters to older users and correlates with high income, fitness level, and usage.  
   - **Recommendation**: Promote KP781 as a premium, durable treadmill for fitness enthusiasts and affluent customers. Use testimonials and success stories from high-profile individuals or fitness influencers.  

---

### **Gender Preferences**
1. Males show stronger preferences overall, particularly for KP781.  
   - **Recommendation**: Focus on tailoring messaging for Male customers by highlighting durability, performance, and advanced features in KP781 campaigns.  

2. Females prefer KP481 and KP281 more evenly.  
   - **Recommendation**: Design campaigns that appeal to Female buyers with aesthetics, affordability, and ease of use as selling points.  

---

### **Marital Status Analysis**
1. Partnered customers consistently make more purchases across all products.  
   - **Recommendation**: Create offers like “Partner Discounts” or “Family Bundles” to increase engagement from this segment.  

2. Singles have fewer purchases but still represent a significant audience for KP281.  
   - **Recommendation**: Develop targeted campaigns for single customers emphasizing self-care, independence, and personal fitness goals.  

---

### **Income and Fitness Insights**
1. **KP781’s correlation with high income and fitness level** indicates its popularity among wealthier, health-conscious individuals.  
   - **Recommendation**: Reinforce its premium image with marketing campaigns that emphasize advanced features and long-term health benefits.  

2. KP281 and KP481 appeal more to lower-income individuals.  
   - **Recommendation**: Highlight affordability and practical benefits in their advertising campaigns to further engage budget-conscious buyers.
